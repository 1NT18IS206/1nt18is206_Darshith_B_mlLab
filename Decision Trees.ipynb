```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn import tree
```

```python
df = pd.read_csv ('zoo.csv', header = None)
print (df)
```

```python
y = df[:][17]
print(y.shape)
```

```python
x = df.iloc[: , 1:17]
```

```python
from sklearn.model_selection import train_test_split

train_x, test_x, train_y, test_y = train_test_split(x, y, test_size = 0.2, random_state = 111)


print(train_x.shape, train_y.shape)
```

```python
model1 = tree.DecisionTreeClassifier(criterion="entropy")
model1 = model1.fit(train_x,train_y)
```

```python
model2 = tree.DecisionTreeClassifier()
model2 = model2.fit(train_x,train_y)
```

```python
plt.figure(figsize = (15,15))
tree.plot_tree(model1)
plt.show()
```

```python
plt.figure(figsize = (15,15))
tree.plot_tree(model2)
plt.show()
```

```python
pred_y1 = model1.predict(test_x)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, cohen_kappa_score
print(confusion_matrix(test_y, pred_y1))
print(classification_report(test_y, pred_y1))
print(cohen_kappa_score(test_y, pred_y1, labels=None, weights=None, sample_weight=None))
```

```python
pred_y2 = model2.predict(test_x)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, cohen_kappa_score
print(confusion_matrix(test_y, pred_y2))
print(classification_report(test_y, pred_y2))
print(cohen_kappa_score(test_y, pred_y2, labels=None, weights=None, sample_weight=None))
```
